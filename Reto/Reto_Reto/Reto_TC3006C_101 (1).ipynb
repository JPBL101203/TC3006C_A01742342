#### **SibSp**

Esta variable nos indica la cantidad de relaciones familiares que tenía una persona, en esta variable se agrupa a los hermanos/parejas. Dentro del análisis es importante saber si esta variable nos puede ser de utilidad ya que podemos ver si el hecho de tener más familiares afecta a la supervivencia de las personas. Por ejemplo, si estas personas pudieron obtener ayuda de sus familiares a la hora de huir del barco o si priorizaron la supervivencia de sus seres queridos.

```python
# Porcentaje de supervivencia en función de SibSp
sibsp_survival = train_data.groupby('SibSp')['Survived'].mean() * 100
sibsp_counts = train_data['SibSp'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
sns.barplot(x=sibsp_survival.index, y=sibsp_survival.values, palette="viridis", hue=sibsp_survival.index, dodge=False, legend=False)
plt.title('Porcentaje de Supervivencia en función de la cantidad de Hermanos/Esposos (SibSp)')
plt.xlabel('Cantidad de Hermanos/Esposos a bordo')
plt.ylabel('Porcentaje de Supervivencia (%)')

# Mostrar el número de pasajeros y el porcentaje de supervivencia por cada categoría
for i, count in enumerate(sibsp_counts):
    plt.text(i, sibsp_survival.iloc[i] + 1, f'{sibsp_survival.iloc[i]:.2f}%', ha='center')
    plt.text(i, sibsp_survival.iloc[i] - 5, f'N={count}', color='white', ha='center')

plt.ylim(0, 60)
plt.show()
```

De la siguiente gráfica podemos observar que sí existe un aumento en el porcentaje de supervivencia en las personas que tengan almenos 1 o más hermanos/parejas, aunque podemos observar un decremento si la cifra sube a más de 2.

## **Transformación**

Como una de nuestras variables significativa (edad) tiene datos faltantes, se buscará imputar estos datos con distintas técnicas. Además, se buscará simplificar las demás variables y convertir hacer un encoding de las variables categoricas significativas.

#### **Mediana**

Imputación de edad por mediana según la "Pclass", "Sex" y el "Title". Se dividen los  datos en grupos según características como la clase social, el género o el título, usando la función groupby en pandas. Después, para cada uno de estos grupos, calculas la mediana de una variable específica, como la edad, con el método .median(). Esto te da una serie de medianas que muestran el valor del medio para cada grupo. Este método es muy útil porque te permite ajustar tus cálculos a las particularidades de cada grupo, lo cual es especialmente valioso cuando necesitas reemplazar valores faltantes en tus datos de manera que reflejen mejor la realidad de cada subgrupo.

```python
train_x_median = train_data.copy()
test_x_median = test_data.copy()

# Calcular medianas por combinación de Pclass, Sex, y Title
median_title = train_x_median.groupby(['Pclass', 'Sex', 'Title'])['Age'].median()

# Calcular medianas por combinación de Pclass y Sex (sin considerar Title)
median_sex = train_x_median.groupby(['Pclass', 'Sex'])['Age'].median()

# Calcular medianas solo por Pclass
median_pclass = train_x_median.groupby('Pclass')['Age'].median()

# Mediana global
median_global = train_x_median['Age'].median()

# Preparar DataFrame para las medianas
median_title_df = median_title.reset_index()
median_title_df.columns = ['Pclass', 'Sex', 'Title', 'Median_Age_Title']

median_sex_df = median_sex.reset_index()
median_sex_df.columns = ['Pclass', 'Sex', 'Median_Age_Sex']

median_pclass_df = median_pclass.reset_index()
median_pclass_df.columns = ['Pclass', 'Median_Age_Pclass']

# Imputación en el dataset de entrenamiento
def age_median_impute(row):
    if pd.isna(row['Age']):
        key_title = (row['Pclass'], row['Sex'], row['Title'])
        key_sex = (row['Pclass'], row['Sex'])
        key_pclass = row['Pclass']

        # Intentar encontrar la mediana por Title
        median = median_title.get(key_title, np.nan)

        # Si no hay mediana por Title, intentar encontrar por Sex
        if pd.isna(median):
            median = median_sex.get(key_sex, np.nan)

        # Si no hay mediana por Sex, intentar encontrar por Pclass
        if pd.isna(median):
            median = median_pclass.get(key_pclass, np.nan)

        # Si no hay mediana por Pclass, usar mediana global
        if pd.isna(median):
            median = median_global

        return median
    else:
        return row['Age']

# Imputar en el dataset de entrenamiento
train_x_median = train_x_median.merge(median_title_df, on=['Pclass', 'Sex', 'Title'], how='left')
train_x_median = train_x_median.merge(median_sex_df, on=['Pclass', 'Sex'], how='left')
train_x_median = train_x_median.merge(median_pclass_df, on='Pclass', how='left')

# Imputar valores faltantes con la mediana
train_x_median['Age'] = train_x_median.apply(age_median_impute, axis=1)

# Eliminar columnas temporales
train_x_median.drop(columns=['Median_Age_Title', 'Median_Age_Sex', 'Median_Age_Pclass'], inplace=True)

# Imputar en el dataset de prueba
test_x_median = test_x_median.merge(median_title_df, on=['Pclass', 'Sex', 'Title'], how='left')
test_x_median = test_x_median.merge(median_sex_df, on=['Pclass', 'Sex'], how='left')
test_x_median = test_x_median.merge(median_pclass_df, on='Pclass', how='left')

# Imputar valores faltantes con la mediana
test_x_median['Age'] = test_x_median.apply(age_median_impute, axis=1)

# Eliminar columnas temporales
test_x_median.drop(columns=['Median_Age_Title', 'Median_Age_Sex', 'Median_Age_Pclass'], inplace=True)

# Verifica los primeros registros
print(train_x_median.head())
print(test_x_median.head())
```

#### **Media**


Imputación de edad por media según la "Pclass", "Sex" y el "Title". Se dividen los datos en grupos según características como la clase social, el género o el título, usando la función groupby en pandas. Luego, para cada uno de estos grupos, calculas la media de una variable específica, como la edad, utilizando el método .mean(). Esto te da una serie de medias que representan el valor promedio de la variable para cada grupo. Este método es útil porque te permite hacer ajustes basados en la generalización de cada grupo, lo que es valioso al reemplazar valores faltantes de manera que reflejen mejor la distribución promedio dentro de cada subgrupo. Sin embargo, a diferencia de la mediana, la media puede ser influenciada por valores extremos, por lo que es importante considerar la naturaleza de los datos antes de utilizarla.

```python
train_x_mean = train_data.copy()
test_x_mean = test_data.copy()

# Calcular la media por combinación de Pclass, Sex, y Title
mean_title = train_x_mean.groupby(['Pclass', 'Sex', 'Title'])['Age'].mean()

# Calcular la media por combinación de Pclass y Sex (sin considerar Title)
mean_sex = train_x_mean.groupby(['Pclass', 'Sex'])['Age'].mean()

# Calcular la media solo por Pclass
mean_pclass = train_x_mean.groupby('Pclass')['Age'].mean()

# Media global
mean_global = train_x_mean['Age'].mean()

# Preparar DataFrames para las medias
mean_title_df = mean_title.reset_index()
mean_title_df.columns = ['Pclass', 'Sex', 'Title', 'Mean_Age_Title']

mean_sex_df = mean_sex.reset_index()
mean_sex_df.columns = ['Pclass', 'Sex', 'Mean_Age_Sex']

mean_pclass_df = mean_pclass.reset_index()
mean_pclass_df.columns = ['Pclass', 'Mean_Age_Pclass']

# Imputación en el dataset de entrenamiento
def age_mean_impute(row):
    if pd.isna(row['Age']):
        key_title = (row['Pclass'], row['Sex'], row['Title'])
        key_sex = (row['Pclass'], row['Sex'])
        key_pclass = row['Pclass']

        # Intentar encontrar la media por Title
        mean = mean_title.get(key_title, np.nan)

        # Si no hay media por Title, intentar encontrar por Sex
        if pd.isna(mean):
            mean = mean_sex.get(key_sex, np.nan)

        # Si no hay media por Sex, intentar encontrar por Pclass
        if pd.isna(mean):
            mean = mean_pclass.get(key_pclass, np.nan)

        # Si no hay media por Pclass, usar media global
        if pd.isna(mean):
            mean = mean_global

        return mean
    else:
        return row['Age']

# Imputar en el dataset de entrenamiento
train_x_mean = train_x_mean.merge(mean_title_df, on=['Pclass', 'Sex', 'Title'], how='left')
train_x_mean = train_x_mean.merge(mean_sex_df, on=['Pclass', 'Sex'], how='left')
train_x_mean = train_x_mean.merge(mean_pclass_df, on='Pclass', how='left')

# Imputar valores faltantes con la media
train_x_mean['Age'] = train_x_mean.apply(age_mean_impute, axis=1)

# Eliminar columnas temporales
columns_to_drop = ['Mean_Age_Title', 'Mean_Age_Sex', 'Mean_Age_Pclass']
train_x_mean.drop(columns=columns_to_drop, inplace=True)

# Imputar en el dataset de prueba
test_x_mean = test_x_mean.merge(mean_title_df, on=['Pclass', 'Sex', 'Title'], how='left')
test_x_mean = test_x_mean.merge(mean_sex_df, on=['Pclass', 'Sex'], how='left')
test_x_mean = test_x_mean.merge(mean_pclass_df, on='Pclass', how='left')

# Imputar valores faltantes con la media
test_x_mean['Age'] = test_x_mean.apply(age_mean_impute, axis=1)

# Eliminar columnas temporales
test_x_mean.drop(columns=columns_to_drop, inplace=True)

# Verifica los primeros registros
print(train_x_mean.head())
print(test_x_mean.head())
```

### **Regresión Logística**

Regresión Logística es un modelo de aprendizaje automático utilizado para problemas de clasificación binaria, aunque también se puede extender a múltiples clases. Este algoritmo estima la probabilidad de que una observación pertenezca a una clase específica mediante una función logística. La función logística transforma una combinación lineal de las características de entrada en un valor entre 0 y 1, que se interpreta como una probabilidad. Si la probabilidad es mayor que un umbral, la observación se clasifica en una clase; de lo contrario, se clasifica en la otra. La regresión logística es eficaz para problemas donde las relaciones entre las características y las clases son aproximadamente lineales y es fácil de interpretar debido a su naturaleza probabilística.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def evaluate_logistic_regression(X_train, X_val, y_train, y_val):
    # Inicializar modelo
    model = LogisticRegression(max_iter=1000, random_state=42)
    # Entrenar modelo
    model.fit(X_train, y_train)
    # Predecir datos con modelo
    y_pred = model.predict(X_val)

    # Calcular exactitud de predicción
    accuracy = accuracy_score(y_val, y_pred)
    # Calcular matriz de confusión
    conf_matrix = confusion_matrix(y_val, y_pred)
    # Generar reporte de clasificación
    class_report = classification_report(y_val, y_pred, target_names=['No Sobrevive', 'Sobrevive'], output_dict=True)

    # Regresar resultados
    return {
          'Método': 'Regresión Logística',
          'Imputación': 'Variable',
          'Exactitud': accuracy,
          'Matriz de Confusión': conf_matrix,
          'Reporte de Clasificación': class_report
      }
```

Se utiliza el modelo de regresión logística de Scikit Learn con hiperparámetros default, a excepción del número máximo de iteraciones que cambiamos a 1000 y el random state que se agregó como 42.
La configuración por defecto incluye la regularización L2, ajusta un intercepto, y utiliza el solver L-BFGS con un máximo de 100 iteraciones. No se ajustan pesos de clase ni se fija una semilla para la aleatoriedad. El modelo funciona tanto para clasificación binaria como multiclase, y usa un solo núcleo de CPU por defecto.

La razón es que ofrece un buen equilibrio entre rendimiento y simplicidad para la mayoría de los problemas. El uso de la regularización L2 y el solver L-BFGS garantiza una solución eficiente y estable sin necesidad de ajustes adicionales, lo que permite obtener resultados rápidos y confiables en problemas estándar de clasificación.

### **Árbol de decisión**

Árbol de Decisión es un algoritmo de aprendizaje automático que utiliza una estructura de árbol para tomar decisiones basadas en características de los datos. El árbol se construye dividiendo repetidamente el conjunto de datos en función de las características que mejor separan las clases o valores en cada paso. Cada nodo del árbol representa una decisión basada en una característica, y cada rama representa un posible valor o intervalo para esa característica. Los nodos terminales, o hojas, contienen la predicción final, ya sea una clase en problemas de clasificación o un valor en problemas de regresión. Los árboles de decisión son fáciles de interpretar y visualizan, pero pueden ser propensos al sobreajuste si no se podan adecuadamente.

```python
from sklearn.tree import DecisionTreeClassifier

def evaluate_decision_tree(X_train, X_val, y_train, y_val):
    # Inicializar modelo
    model = DecisionTreeClassifier(random_state=42)
    # Entrenar modelo
    model.fit(X_train, y_train)
    # Predecir datos con modelo
    y_pred = model.predict(X_val)

    # Calcular exactitud de predicción
    accuracy = accuracy_score(y_val, y_pred)
    # Calcular matriz de confusión
    conf_matrix = confusion_matrix(y_val, y_pred)
    # Generar reporte de clasificación
    class_report = classification_report(y_val, y_pred, target_names=['No Sobrevive', 'Sobrevive'], output_dict=True)

    # Regresar resultados
    return {
          'Método': 'Árbol de Decisión',
          'Imputación': 'Variable',
          'Exactitud': accuracy,
          'Matriz de Confusión': conf_matrix,
          'Reporte de Clasificación': class_report
      }
```

Se utiliza el modelo de Árbol de Decisión de Scikit Learn con hiperparámetros default, a excepción del random state que se agregó como 42. Por defecto, usa el criterio Gini para dividir los nodos y no limita la profundidad del árbol. Las divisiones se hacen con al menos 2 muestras y las hojas pueden tener solo 1 muestra. No se ajustan los pesos de clase ni se realiza poda, y todas las características son consideradas en cada división.

Utilizamos la configuración por defecto ya que permite crear árboles complejos que se ajustan bien a los datos sin necesidad de hacer ajustes manuales. Esto facilita la captura de relaciones no lineales y es adecuado para problemas donde se busca un alto nivel de detalle en las predicciones.

### **Bosque Aleatorio**

Bosque Aleatorio es un algoritmo de aprendizaje automático que utiliza un conjunto de árboles de decisión para mejorar la precisión y la robustez de las predicciones. Cada árbol en el "bosque" se entrena con una muestra aleatoria de los datos y un subconjunto aleatorio de características, lo que introduce diversidad y reduce la correlación entre los árboles. Para hacer una predicción, el algoritmo combina las decisiones de todos los árboles, ya sea mediante votación en problemas de clasificación o promediando en problemas de regresión. Esta técnica ayuda a minimizar el sobreajuste y puede proporcionar estimaciones sobre la importancia de las características en el modelo.

```python
from sklearn.ensemble import RandomForestClassifier

def evaluate_random_forest(X_train, X_val, y_train, y_val):
    # Inicializar modelo
    model = RandomForestClassifier(random_state=42)
   # Entrenar modelo
    model.fit(X_train, y_train)
    # Predecir datos con modelo
    y_pred = model.predict(X_val)

    # Calcular exactitud de predicción
    accuracy = accuracy_score(y_val, y_pred)
    # Calcular matriz de confusión
    conf_matrix = confusion_matrix(y_val, y_pred)
    # Generar reporte de clasificación

    class_report = classification_report(y_val, y_pred, target_names=['No Sobrevive', 'Sobrevive'], output_dict=True)

    # Regresar resultados
    return {
          'Método': 'Bosque Aleatorio',
          'Imputación': 'Variable',
          'Exactitud': accuracy,
          'Matriz de Confusión': conf_matrix,
          'Reporte de Clasificación': class_report
      }
```

Se utiliza el modelo de Random Forest de Scikit Learn con hiperparámetros default, a excepción del random state que se agregó como 42. Por defecto, crea 100 árboles que se entrenan usando la técnica de muestreo con reemplazo (bootstrap). Para clasificación, usa el criterio de Gini. Al construir los árboles, selecciona características de manera aleatoria, considerando solo una fracción de ellas (la raíz cuadrada del total en clasificación y todas en regresión). No limita la profundidad de los árboles, permitiendo que crezcan hasta que no se puedan dividir más. También permite dividir nodos con al menos 2 muestras y formar hojas con al menos 1 muestra. Todas las clases tienen el mismo peso, y por defecto, no se usa procesamiento en paralelo.

El modelo por defecto ofrece un rendimiento sólido en una gran variedad de problemas sin necesidad de ajustar los hiperparámetros. Esto nos permite obtener un modelo robusto y fiable rápidamente, lo que nos fue especialmente útil para la situación en que no se tiene tiempo o recursos para realizar una optimización exhaustiva de hiperparámetros.

### **K-nearest neighbours**

k-Nearest Neighbors (k-NN) es un algoritmo de aprendizaje automático simple y efectivo utilizado para clasificación y regresión. Funciona asignando una etiqueta o valor a un punto de datos basándose en las etiquetas o valores de sus k vecinos más cercanos en el espacio de características. Para la clasificación, el algoritmo predice la clase del punto de datos en función de la mayoría de las clases de sus k vecinos más cercanos. Para la regresión, predice el valor como el promedio de los valores de sus k vecinos más cercanos. La proximidad entre puntos se mide típicamente usando distancias como la Euclidiana. k-NN es intuitivo y fácil de implementar, pero su rendimiento puede verse afectado por la elección de k y la escala de las características.

```python
from sklearn.neighbors import KNeighborsClassifier

def evaluate_knn(X_train, X_val, y_train, y_val):
    # Inicializar modelo
    model = KNeighborsClassifier()
    # Entrenar modelo
    model.fit(X_train, y_train)
    # Predecir datos con modelo
    y_pred = model.predict(X_val)

    # Calcular exactitud de predicción
    accuracy = accuracy_score(y_val, y_pred)
    # Calcular matriz de confusión
    conf_matrix = confusion_matrix(y_val, y_pred)
    # Generar reporte de clasificación
    class_report = classification_report(y_val, y_pred, target_names=['No Sobrevive', 'Sobrevive'], output_dict=True)

    # Regresar resultados
    return {
          'Método': 'K-NN',
          'Imputación': 'Variable',
          'Exactitud': accuracy,
          'Matriz de Confusión': conf_matrix,
          'Reporte de Clasificación': class_report
      }
```

Se utiliza el modelo de K-nearest neighbours de Scikit Learn con hiperparámetros default. Por defecto, utiliza 5 vecinos más cercanos para hacer predicciones. Todos los vecinos tienen el mismo peso en la decisión, y la distancia se calcula utilizando la distancia Euclidiana. El algoritmo que se usa para encontrar los vecinos se selecciona automáticamente según los datos.

Dicho modelo por defecto, sin necesidad de ajustar hiperparámetros, ofrece una buena solución inicial para el problema, permitiendo obtener resultados rápidos sin complicaciones.

## **Evaluación**

El siguiente bloque de código evalúa los modelos. Se crea un diccionario con los métodos de imputación de los datos para ir iterando el tipo de imputación y evaluarlo con cada uno de los modelos.  Primero, separa las características (X) y la variable objetivo (y) y divide los datos en conjuntos de entrenamiento y validación. Se mandan los datos al modelo creado. El modelo realiza predicciones sobre los datos de validación. Finalmente, evalúa el modelo calculando la precisión, la matriz de confusión y un informe de clasificación, proporcionando una visión detallada del rendimiento del modelo.

```python
# Lista para almacenar los resultados
results = []

# Técnicas de imputación y sus correspondientes DataFrames
imputation_methods = {
    'Mediana': train_x_median,
    'Media': train_x_mean,
    '1 Std': train_x_1std,
    '2 Std': train_x_2std
}


# Evaluar cada técnica de imputación con cada modelo
for method_name, method_data in imputation_methods.items():
    # Preparar los datos
    X_train, X_val, y_train, y_val = train_test_split(method_data, train_y_real, test_size=0.2, random_state=42)
    y_train = y_train.values.ravel()
    y_val = y_val.values.ravel()

    # Evaluar regresión logística
    results.append(evaluate_logistic_regression(X_train, X_val, y_train, y_val))
    results[-1]['Imputación'] = method_name

    # Evaluar árbol de decisión
    results.append(evaluate_decision_tree(X_train, X_val, y_train, y_val))
    results[-1]['Imputación'] = method_name

    # Evaluar random forest
    results.append(evaluate_random_forest(X_train, X_val, y_train, y_val))
    results[-1]['Imputación'] = method_name

    # Evaluar k-NN
    results.append(evaluate_knn(X_train, X_val, y_train, y_val))
    results[-1]['Imputación'] = method_name

# Convertir la lista de resultados en un DataFrame
results_df = pd.DataFrame(results)

# Mostrar el DataFrame con los resultados
results_df.head()
```

```python
def parse_classification_report(report_dict):
    """
    Extrae las métricas de precisión, recall y F1-Score para ambas clases.

    :param report_dict: Diccionario del reporte de clasificación
    :return: Serie de pandas con precision, recall y f1-score para cada clase
    """
    metrics = {
        'No Sobrevive': {'precision': float('nan'), 'recall': float('nan'), 'f1-score': float('nan')},
        'Sobrevive': {'precision': float('nan'), 'recall': float('nan'), 'f1-score': float('nan')}
    }

    # Convertir el reporte de clasificación a un diccionario si no lo es
    if isinstance(report_dict, str):
        report_dict = eval(report_dict)  # Evaluar el string como diccionario

    # Extraer métricas para las clases 'No Sobrevive' y 'Sobrevive'
    if 'No Sobrevive' in report_dict:
        metrics['No Sobrevive']['precision'] = report_dict['No Sobrevive']['precision']
        metrics['No Sobrevive']['recall'] = report_dict['No Sobrevive']['recall']
        metrics['No Sobrevive']['f1-score'] = report_dict['No Sobrevive']['f1-score']
    if 'Sobrevive' in report_dict:
        metrics['Sobrevive']['precision'] = report_dict['Sobrevive']['precision']
        metrics['Sobrevive']['recall'] = report_dict['Sobrevive']['recall']
        metrics['Sobrevive']['f1-score'] = report_dict['Sobrevive']['f1-score']

    # Convertir el diccionario de métricas en una Serie de pandas
    return pd.Series({
        'No Sobrevive Precision': metrics['No Sobrevive']['precision'],
        'No Sobrevive Recall': metrics['No Sobrevive']['recall'],
        'No Sobrevive F1-Score': metrics['No Sobrevive']['f1-score'],
        'Sobrevive Precision': metrics['Sobrevive']['precision'],
        'Sobrevive Recall': metrics['Sobrevive']['recall'],
        'Sobrevive F1-Score': metrics['Sobrevive']['f1-score']
    })

    # Aplicar la función a cada fila del DataFrame
classification_metrics_df = results_df['Reporte de Clasificación'].apply(parse_classification_report)

# Concatenar las métricas de clasificación con el DataFrame original
results_df = pd.concat([results_df, classification_metrics_df], axis=1)

classification_metrics_df.head()
```

## **Matriz de Confusión**

Una matriz de confusión es una herramienta de evaluación utilizada para medir el desempeño de un modelo de clasificación. Organiza y presenta los resultados de las predicciones realizadas por el modelo de manera tabular, comparando las predicciones con las etiquetas reales. Es especialmente útil en problemas de clasificación binaria o multiclase, ya que proporciona una visión más detallada de los errores cometidos por el modelo.

```python
 from sklearn.metrics import ConfusionMatrixDisplay

# Configura el estilo de las gráficas
plt.style.use('seaborn-whitegrid')

# Ajusta el tamaño de la figura y el espaciado
fig, axes = plt.subplots(len(results_df), 1, figsize=(4, 4 * len(results_df)), squeeze=False)
plt.subplots_adjust(hspace=0.5)

# Graficar la matriz de confusión para cada modelo e imputación
for i, (idx, result) in enumerate(results_df.iterrows()):
    ax = axes[i, 0]
    cm = result['Matriz de Confusión']
    ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Sobrevive', 'Sobrevive']).plot(ax=ax, colorbar=False)
    ax.set_title(f'{result["Método"]} - Imputación: {result["Imputación"]}')
    ax.set_xlabel('')
    ax.set_ylabel('')

# Ajustar el diseño para evitar superposición
plt.tight_layout()
plt.show()
```

En la siguiente serie de matrices de confusión se observa en todas el mismo patrón, en las que las áreas más claras indican una mayor cantidad de incidencias, para todas las matrices el cuadrante de los verdaderos negativos es la más clara (con más observaciones) y es consistente que la segunda más clara es el cuadrante de los verdaderos positivos. Esto podría indicar que hay un pequeño desajuste en el balance de clases. Sin embargo, parece ser que todos los modelos poseen una alta especialidad, lo que quiere decir que son muy buenos en predecir correctamente los casos negativos sin confundirlos con los positivos.

## **Conclusiones Finales**

El Random Forest con imputación de la mediana fue el mejor modelo en este caso debido a su capacidad para manejar tanto la heterogeneidad de las características como los valores faltantes de manera eficaz. A diferencia de modelos más simples como el K-Nearest Neighbors (K-NN) o la regresión logística, Random Forest tiene la ventaja de ser un ensamble de múltiples árboles de decisión, lo que le permite capturar relaciones no lineales y complejas entre las variables del conjunto de datos del Titanic, como la interacción entre Pclass, Sex, Age y otros atributos. La imputación por mediana, por su parte, es una técnica robusta ante la presencia de outliers y distribuciones asimétricas, lo que ayuda a preservar la estabilidad del modelo al rellenar valores faltantes en la variable Age. Este enfoque es especialmente útil en este dataset, donde la edad varía ampliamente entre los pasajeros y está correlacionada con su probabilidad de supervivencia. Por otro lado, métodos como K-NN tienden a ser menos efectivos en datasets con variables de diferentes tipos (categóricas y numéricas), y la regresión logística, al ser un modelo lineal, no captura de manera adecuada las complejas relaciones entre las variables, lo que limita su rendimiento en comparación con Random Forest. En conjunto, la combinación de la robustez de Random Forest y la estabilidad de la imputación por mediana lo convierte en el enfoque más adecuado para este problema.

Finalmente, es importante recalcar que este análisis de supervivencia realza características que nos dan información acerca de los estereotipos, según el contexto histórico, seguidos por los pasajeros para explicar porqué se salvaron los pasajeros que se salvaron en el RMS Titanic. Esta es información que también nos indica que este modelo no es utilizable para ningún otro caso, pues el accidente es irreproducible y dichos estereotipos han cambiado a lo largo de los 112 años posteriores a la tragedia.
